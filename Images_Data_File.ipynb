```python
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import requests
import time
```

Imported necessary libraries

```python
driver = webdriver.Chrome(r'chromedriver.exe')
```

Using amazon.in to scrap the data

```python
input_img_1 = str(input('Search Image 1 = ' ))
input_img_2 = str(input('Search Image 2 = ' ))
input_img_3 = str(input('Search Image 3 = ' ))
```

```python
urls_saree = [] # Scrapping href links for sarees

url = 'https://www.amazon.in/'
driver.get(url)    
    

search = driver.find_element(By.XPATH,'//input[@id="twotabsearchtextbox"]')
search.send_keys(input_img_1)

search_button = driver.find_element(By.XPATH,'//input[@id="nav-search-submit-button"]')
search_button.click()

time.sleep(4)

url = driver.find_elements(By.XPATH,'//a[@class="s-pagination-item s-pagination-button"]')
urls_saree.append(driver.current_url)
for i in url:
    urls_saree.append(i.get_attribute("href"))
```

```python
urls_trouser = [] # Scrapping href links for men trouser

url = 'https://www.amazon.in/'
driver.get(url)    
    
    
search = driver.find_element(By.XPATH,'//input[@id="twotabsearchtextbox"]')
search.send_keys(input_img_2)

search_button = driver.find_element(By.XPATH,'//input[@id="nav-search-submit-button"]')
search_button.click()

time.sleep(4)

url = driver.find_elements(By.XPATH,'//a[@class="s-pagination-item s-pagination-button"]')
urls_trouser.append(driver.current_url)
for i in url:
    urls_trouser.append(i.get_attribute("href"))
```

```python
urls_jeans = [] # Scrapping href links for men jeans

url = 'https://www.amazon.in/'
driver.get(url)    
    

search = driver.find_element(By.XPATH,'//input[@id="twotabsearchtextbox"]')
search.send_keys(input_img_3)

search_button = driver.find_element(By.XPATH,'//input[@id="nav-search-submit-button"]')
search_button.click()

time.sleep(4)

url = driver.find_elements(By.XPATH,'//a[@class="s-pagination-item s-pagination-button"]')
urls_jeans.append(driver.current_url)
for i in url:
    urls_jeans.append(i.get_attribute("href"))
```

```python
urls_jeans.append('https://www.amazon.in/s?k=men+jeans&page=4') # added one extra link manually 
urls_trouser.append('https://www.amazon.in/s?k=men+trouser&page=4') # added one extra link manually
urls_saree.append('https://www.amazon.in/s?k=saree&page=4') # added one extra link manually
```

```python
image_urls_saree = [] # Collecting Images Link for Sarees

for each_url in urls_saree:
    driver.get(each_url) 
    time.sleep(2)
    images = driver.find_elements(By.XPATH,'//img[@class="s-image"]')
    for io in images:
        image_urls_saree.append(io.get_attribute("src"))
```

```python
image_urls_trouser = [] # Collecting Images Link for Trouser

for each_url in urls_trouser:
    driver.get(each_url) 
    time.sleep(2)
    images = driver.find_elements(By.XPATH,'//img[@class="s-image"]')
    for io in images:
        image_urls_trouser.append(io.get_attribute("src"))
```

```python
image_urls_jeans = [] # Collecting Images Link for Jeans

for each_url in urls_jeans:
    driver.get(each_url) 
    time.sleep(2)
    images = driver.find_elements(By.XPATH,'//img[@class="s-image"]')
    for io in images:
        image_urls_jeans.append(io.get_attribute("src"))
```

Collected URLs to scrap the data

```python
print (len(image_urls_saree))
print (len(image_urls_trouser))
print (len(image_urls_jeans))
```

```python
for img in image_urls_saree: # Downloaded Saree Images
    img_file = img.split('/')[-1]
    print(img_file, "- Done")
    r = requests.get(img, stream=True)
    if r.status_code == 200:
        with open(img_file, 'wb') as f:
            for u in r:
                f.write(u)
                
    else:
        print('')
```

```python
for img in image_urls_trouser: # Downloaded trouser Images
    img_file = img.split('/')[-1]
    print(img_file, "- Done")
    r = requests.get(img, stream=True)
    if r.status_code == 200:
        with open(img_file, 'wb') as f:
            for u in r:
                f.write(u)
                
    else:
        print('')
```

```python
for img in image_urls_jeans: # Downloaded Jeans Images
    img_file = img.split('/')[-1]
    print(img_file, "- Done")
    r = requests.get(img, stream=True)
    if r.status_code == 200:
        with open(img_file, 'wb') as f:
            for u in r:
                f.write(u)
                
    else:
        print('')
```
